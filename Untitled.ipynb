```python
creditcard = pd.read_csv('creditcard.csv')
demographics = pd.read_csv('demographics.csv')
creditcardtrx = pd.read_csv('creditcardtrx.csv')
depositaccount = pd.read_csv('depositaccount.csv')
loanaccount = pd.read_csv('loanaccount.csv')
savingsaccount = pd.read_csv('savingsaccount.csv')
```

```python
from functools import reduce

dfs = [demographics,creditcard, savingsaccount, depositaccount]
df_final = reduce(lambda left,right: pd.merge(left,right, how='left'), dfs)
```

```python
df_final.head()
```

```python
df_final['Type'].fillna('No', inplace=True)
```

```python
from sklearn.preprocessing import LabelEncoder

lb_make = LabelEncoder()
df_final["card_type"] = lb_make.fit_transform(df_final["Type"])
df_final[["Type", "card_type"]].head(11)
```

```python
df_final["Gender"] = lb_make.fit_transform(df_final["gender"])
df_final[["gender", "Gender"]].head(11)
```

```python
df_final["Occupation"] = lb_make.fit_transform(df_final["occupation"])
df_final[["Oender", "occupation"]].head(11)
```
